```python
import pandas as pd
import numpy as np
# 2. Load the file created by the team
# Note: Ensure this file exists in the folder
file_path = f'{BASE}/task3-combination/task3_supervised_output.csv'
df_task3 = pd.read_csv(file_path)
```

```python
# Shirin's signal phrases
signals = [
    "should", "could", "wish", "needs", "missing",
    "would be better", "hope", "add more", "next time", "recommend"
]

# Create regex pattern
pattern = '|'.join(signals)

# Filter for rows containing these signals
# We use 'comments' because 'cleaned_text' might have removed these words
suggestions_df = df_task3[df_task3['comments'].str.contains(pattern, case=False, na=False)].copy()

print(f"✅ Found {len(suggestions_df):,} suggestion-related reviews.")
```

```python
# Select relevant columns
# 'predicted_aspect' was generated by Williams/Boone in the previous step
suggestions_final = suggestions_df[['listing_id', 'predicted_aspect', 'comments']]

# Get top 3 examples per listing per aspect to keep the dashboard snappy
top_suggestions = suggestions_final.groupby(['listing_id', 'predicted_aspect']).head(3)

# Export to the shared folder
output_path = f'{BASE}/task3-combination/suggestions_by_aspect.csv'
top_suggestions.to_csv(output_path, index=False)

print(f"✅ Done! 'suggestions_by_aspect.csv' is ready at: {output_path}")
```
